## Que es un dato atípico?

Son aquellos datos que presentan algunas caracteristicas que acontinuación se desglosan.

### Caracteristicas

- Se desvían del comportamiento normal/conocido de los datos.
- Son valores que estan lejos de los valores esperados (media).
- No esta conectado a ningún otro dato en cuantos a sus caracteristicas.

Por tanto, los valores atípicos generalmente exhiben alguna anormalidad o algún tipo de comportamiento fuera del conjunto de datos y pueden alterar una función positiva o negativamente. 

Aunque existen muchas definiciones para el outlier, estas estan en función de las caracteristicas observadas del conjunto de datos. Una de ellas establece que el valor atípico es un objeto que se desvía tanto de otros como para despertar sospechas de que es generado de una fuente diferente. Esta definción capta la gravedad de la desviación de un objeto y otorga una impresión de que dicho objeto es completamente diferente a los demás.

### Otros nombres para valor atípico

- Atípico
- Anormal
- Valor Extremo
- Valor anómalo




## Caracterización de los Outliers

El comportamiento de la desviación de los valores atípicos se puede atribuir a varias propiedades observadas en los datos. Cada una de estas propiedades puede conducir a una forma específica de caracterización de los valores atípicos, como se explica a continuación.


### Basado en el Tamaño

Es posible tener valores atípicos de tamaños variados según la cantidad de objetos involucrados.

- un solo objeto.
- un pequeño subconjunto de objetos.
- una comunidad/grupo de objetos.
- una red/gráfico completo.

### Basado en la Diversidad

En el caso de una recopilación de datos distribuida normalmente, podría ser un objeto que se sitúa lejos de la media en función de la estructura de varianza/covarianza. Por ejemplo, en una distribución normal univariada, los valores en el rango **[μ − 3σ, μ + 3σ]**, representan el **99,73 %**. Por tanto, cualquier objeto/valor que se encuentre fuera de este rango puede calificar para ser un valor atípico, ya que está alejado de la media.

### Basado en Roles

Comúnmente se percibe que los valores atípicos tienen un impacto indeseable o negativo en la actividad de procesamiento de datos. Sin embargo, hay varias situaciones en las que los valores atípicos pueden desempeñar un papel positivo.

1. En una comunidad de los elementos sociales, es posible tener una persona/entidad demasiado central o influyente. Tal persona resulta ser un caso atípico. Sin embargo, con el fin de promocionar un producto, la persona influyente puede desempeñar un papel fundamental, ya que puede influir positivamente en una gran parte de las personas de su comunidad.

2. En la clasificación basada en SVM, el clasificador aprende de un subconjunto del conjunto de patrones de entrenamiento. Aquí, el aprendizaje está destinado a ajustar el vector de peso de la función discriminante (lineal) asociada con la SVM. Todos los demás patrones (vectores que no son de soporte) se descartan o ignoran en este proceso. Por tanto, los valores atípicos son útiles para hacer que el clasificador aprenda los patrones utilizando los datos del subconjunto.


## Importancia de los Outliers

Es importante distinguir los valores atípicos del ruido aleatorio que normalmente se observa en los datos de la vida real, ya que ambos tipos de datos se caracterizan por su desviación de los datos normales. Una desviación suficiente para que un objeto se considere atípico suele ser un juicio subjetivo basado en el interés del analista.
Si bien el ruido en sí mismo puede no ser interesante para el analista, su identificación y eliminación sigue siendo una importante tarea de minería de datos.

Por tanto, los problemas de detección de ruido y de valores atípicos son igualmente importantes para un análisis de datos eficaz. En este contexto, la mejor manera de encontrar valores atípicos y distinguirlos del ruido es utilizar la retroalimentación de ejemplos de interés de valores atípicos previamente conocidos.

En escenarios no supervisados, el ruido representa el límite semántico entre los datos normales y los verdaderos valores atípicos. Por tanto, se modela como una forma débil de valores atípicos. 

Desde una perspectiva teórica, se espera que los algoritmos de detección supervisada sean más eficaz, ya que las características de los ejemplos conocidos se pueden utilizar para afinar el proceso de búsqueda hacia valores atípicos más relevantes. 
Por lo tanto, el nucleo del problema es que los valores atípicos deben ser inusuales de una manera interesante.

Si bien los métodos no supervisados se pueden utilizar para la eliminación de ruido o la detección de valores atípicos, los métodos supervisados son más adecuados para la detección de valores atípicos específicos de la aplicación.


## Una mirada teórica

Muchas aplicaciones del mundo real se enfrentan al problema de identificar eventos raros (complejos), que se caracterizan por combinaciones contextuales de acciones específicas, a menudo realizadas por múltiples agentes.

Básicamente, esto implica dos tareas importantes de exploración de datos: 


1.  Descubrir tipos de eventos previamente desconocidos.
2.  Detectar instancias de estos tipos de eventos en los datos.


Por tanto, existe una transformación de valores atípicos para eventos de interés.

- Eventos de Interés en un dominio
- Anomalías específicas del dominio
- Valores atípicos descubiertos en los datos



Esto se debe a que los eventos de interés en una aplicación no se caracterizan por su rareza sino por su inexplicabilidad en términos de comportamientos del mundo real aplicables (se conocen a priori).
En tales escenarios, las técnicas de detección de valores atípicos actúan como puntos de partida para un proceso de descubrimiento más amplio.
Por ejemplo, un evento puede ser benigno (simplemente una coincidencia) o malicioso según la explicación y el contexto.

De acuerdo con esta perspectiva, se pueden pensar en los siguientes niveles de abstracción con respecto a diferentes tipos de eventos

**Outliers**: datos inusuales que son estadísticamente raros pero que pueden o no resultar de los mismos procesos generativos.

**Anomalías**:  datos que han resultado de un proceso generativo **distinto** que puede o no aparecer en los datos como valores atípicos.

**Eventos de interés**: Anomalías causadas por algún tipo de comportamiento en el mundo real que preocupa a la aplicación.


Es importante señalar que los valores atípicos siempre se definen con respecto a alguna población de referencia. Básicamente, es una evaluación relativa de los objetos dentro de un conjunto. Además, un valor atípico debe describirse no solo en términos de su probabilidad, sino también en términos que un analista pueda entender para determinar si ese valor atípico es significativo en un determinado contexto.

Por tanto, el conjunto de anomalías identificadas (A) específicas de un dominio de aplicación resulta ser un subconjunto de los valores atípicos descubiertos en los datos correspondientes (O), es decir, A ⊂ O. 

Asimismo, determinar los eventos de interés a partir del conjunto de anomalías requiere una evaluación subjetiva basada en el conocimiento del dominio.


Como se definió anteriormente, los valores atípicos representan información valiosa sobre la naturaleza evolutiva del fenómeno físico subyacente. **Deben verse como indicadores tempranos del cambio potencial en el comportamiento del sistema**. Si el cambio percibido a través de la detección de valores atípicos resulta ser desfavorable (malicioso), entonces se debe tener a mano un plan de mitigación. 

Por tanto, se debe explorar varios aspectos computacionales para la detección de valores atípicos y asi determinar un método adecuado en un contexto de aplicación dado.



## Detección
La detección precisa y eficiente de valores atípicos presentes en un conjunto de datos siempre es un desafío, ya que no existe una definición universal que se adscriba a los valores atípicos.
Como resultado, depende de la persona que se ocupa de este problema considerar una definición adecuada y desarrollar un método aceptable para abordar los requisitos.

Dado el entendimiento anterior, el problema de detección de valores atípicos se puede definir ampliamente como: dado un conjunto de objetos de datos, encontrar una cantidad específica de objetos que son considerablemente diferentes, excepcionales e inconsistentes con respecto a los datos restantes. Esta definición básicamente captura formas distintas de un objeto que muestra inusual
características.

Muchos algoritmos de minería de datos intentan minimizar la influencia de los valores atípicos o eliminarlos por completo. Sin embargo, esto podría resultar en la pérdida de información oculta importante debido al hecho bien conocido de que el ruido de una persona podría ser la señal de otra persona.

Por lo tanto, los propios valores atípicos pueden ser de particular interés, como en el caso de la detección de fraude, donde pueden indicar alguna actividad fraudulenta. Además de la detección de fraudes en aplicaciones financieras, el marketing de nichos y la detección de anomalías en la red son otras aplicaciones destacadas de la detección de valores atípicos, lo que la convierte en una importante tarea de minería de datos.

### Método basado en la Distancia


 Un dato se trata como un valor atípico si no hay más de **k** objetos en el conjunto de datos a una distancia de **d** o menos de ese objeto.
Esta forma de encontrar valores atípicos no necesita el conocimiento previo de la distribución de datos que requieren los métodos estadísticos. Sin embargo, este método requiere que el usuario especifique un umbral de distancia **d** y un valor para el parámetro **k** para determinar valores atípicos.
Dentro de estos métodos tenemos los siguientes algoritmos:

- **Distancia al k-ésimo vecino más cercano**: Esencialmente, uno puede clasificar cada objeto en función de su distancia a su k-ésimo vecino más cercano, y luego considerar un número deseado de valores atípicos mejor clasificados de esta lista de clasificados.

- **Distancia promedio a los k-vecinos más cercanos**: similar al método anterior, uno puede producir una lista clasificada de objetos utilizando este criterio para medir la distancia y encontrar valores atípicos del número deseado.

- **Suma de las distancias a sus k-vecinos más cercanos**: La suma calculada para cada objeto se asigna como su peso. Los objetos de datos con pesos elevados se consideran valores atípicos.


Todas las técnicas anteriores se basan en la identificación de los vecinos más cercanos de un objeto de datos para determinar los objetos con solo unos pocos vecinos como valores atípicos. Aunque el principio subyacente es el mismo, el conjunto de valores atípicos generados mediante estas técnicas podría ser considerablemente diferente.

Las ventajas de los métodos basados en la distancia son que no es necesario definir una distribución explícita para determinar lo inusual y que se pueden aplicar a cualquier espacio de características para el que se pueda definir una medida de distancia. 
Por lo tanto, los métodos basados en la distancia encuentran aplicabilidad en varios contextos como métodos de facto para la detección de valores atípicos.
Sin embargo, es posible que estos métodos no muestren resultados óptimos cuando los datos de entrada tienen densidades variables locales.


### Método basado en la Densidad

Los enfoques basados en la densidad para la detección de valores atípicos se desarrollan para tratar los problemas de densidad local de los métodos basados en la distancia. En un marco típico basado en la densidad, hay dos parámetros que definen la noción de densidad de un objeto.

- Parámetro MinPts que especifica el número mínimo de objetos
- Parámetro que especifica el volumen bajo consideración.

Estos dos parámetros determinan un umbral de densidad para que funcione el algoritmo. Para detectar valores atípicos basados en la densidad, es necesario comparar las densidades de diferentes conjuntos de objetos en los datos.

El factor de valor atípico local (LOF) es un enfoque de detección de valores atípicos basado en la densidad que emplea la densidad de accesibilidad local (LRD) de los objetos de datos como medida de densidad.

El LRD de un objeto **p**, indicado como LRDMinPts **(p)**, viene dado por el inverso de la distancia de accesibilidad promedio basada en los MinPts-vecinos más cercanos de ese objeto.

Por lo tanto, el LOF de un objeto **p** se calcula como el promedio de la relación entre el LRD de sus vecinos más cercanos (NMinPts(**p**)) y el del objeto.

El cálculo de los valores LOF para todos los objetos en un conjunto de datos requiere una gran cantidad de búsquedas de k-vecinos más cercanos y puede ser computacionalmente costoso.
Para mejorar este aspecto, muchas variantes del método LOF están disponibles con ciertas modificaciones a la caracterización de la densidad del objeto.

1. **Uso de microclústeres (micro-clusters)**: un microclúster MC(n, c, r) es una representación resumida de un grupo de n datos que probablemente pertenezcan al mismo clúster. Aquí, **c** es la media/centro y **r** es el radio del grupo.

2. **Factor de desviación de granularidad múltiple (multi-granularity DEviation Factor)** (MDEF): cualquier dato cuyo valor MDEF se desvíe mucho de los promedios locales se puede considerar como un valor atípico. Sea el **r-vecindario** de un objeto **Pi** el conjunto de objetos dentro de la distancia **r** de **Pi**. Entonces, la MDEF en el radio **r** de un objeto **Pi** es la desviación relativa de la densidad de su vecindad local de la densidad de vecindad local promedio en su vecindad. Por tanto, un objeto cuya densidad de vecindad coincida con el promedio la densidad del vecindario local tendrá un MDEF de 0. Por el contrario, los valores atípicos tendrán MDEF lejos de 0. La medida MDEF hace frente a las variaciones de densidad local en el espacio de características y puede detectar tanto valores atípicos aislados como grupos atípicos.

3. **Basado en distribución local (Local distribution based)**: este enfoque también está diseñado para encontrar valores atípicos locales a partir de la distribución local de los datos de interés. Se pueden extraer 3 caracteristicas locales: distancia promedio local, densidad local y el grado de asimetría local. Los valores atípicos locales se detectan desde el punto de vista de la distribución local caracterizada por estas características. Aunque es similar a LOF, el uso del grado de asimetría local mejora el poder de discriminación de LOF en la detección de valores atípicos.


### Método basado en Clustering

Estos métodos estan motivados por el desequilibrio en la distribución de los valores atípicos frente a los objetos normales en un conjunto de datos. Este desequilibrio prevalece en los datos, ya que los valores atípicos normalmente corresponden solo a una pequeña fracción de los datos totales. Al abordar este requisito, la detección de valores atípicos basada en agrupamiento tiende a considerar los grupos de tamaños pequeños como valores atípicos agrupados. 
La ventaja de este método es que no tienen que ser supervisados. Además, estas técnicas se pueden usar en un modo incremental, es decir, después de establecer los grupos, se pueden insertar nuevos objetos de datos en el sistema y probar los valores atípicos.

Aunque muchos de los algoritmos de agrupación son capaces de encontrar valores atípicos en los datos, no pueden proporcionar una evaluación del grado de desviación/puntuación de los objetos. Por tanto, se requiere ampliar las técnicas básicas de agrupación para manejar la tarea de detección de valores atípicos mediante la incorporación de esquemas de puntuación adecuados para medir la desviación de los objetos. 
Según la definición, cada valor atípico basado en clúster debe ser local para algún clúster específico en la estructura de agrupación. A continuación se presentan algunas de las técnicas basadas en Clustering.

1. **Agrupamiento de k-medias modificado (k-means clustering)**: este es un método de dos fases que identifica pequeños grupos como valores atípicos utilizando árboles de expansión mínimos. Sin embargo, con este método no se dispone de una medida para identificar el grado en que cada objeto es un valor atípico. Además, con este algoritmo no se incorpora el procedimiento para distinguir los pequeños clústeres del resto.

2. **Factor de valor atípico local basado en clústeres (Cluster-Based Local Outlier Factor) (CBLOF)**: este método determina los valores atípicos locales mediante el establecimiento de una estructura de agrupamiento mediante una técnica de un solo paso. El CBLOF de un objeto se mide considerando tanto el tamaño del grupo al que pertenece el objeto como la distancia entre el objeto y su grupo grande más cercano.

3. **Agrupamiento jerárquico de enlace único (Single linkage hierarchical clustering)**: método que funciona según el enfoque de reducción de valores atípicos autoconsistente. Este método puede tratar con conjuntos de datos que comprenden grupos de dilución, es decir, grupos que poseen objetos dispersos en su periferia.

4. **Métricas de cuantización y distancia implícita (Quantization and implied distance metrics)**: este es un método lineal para la detección de valores atípicos que incluye un procedimiento de cuantificación directa y descubre clústeres atípicos de forma explícita. Requiere solo dos escaneos secuenciales para analizar conjuntos de datos residentes, ya que no realiza comparaciones a nivel de objeto y solo almacena los recuentos de celdas.

5. **Agrupación subespacial basada en la densidad (Density-based subspace clustering)**: esta técnica que define valores atípicos con respecto a las agrupaciones subespaciales máximas y no redundantes. Básicamente, el agrupamiento de subespacios tiene como objetivo encontrar grupos en cualquier subespacio de un espacio de características de alta dimensión y, por lo tanto, puede dar como resultado algunos grupos superpuestos en diferentes subespacios.

6. **Agrupación espacial basada en la densidad local (Local density based spatial clustering)**: este método da importancia al comportamiento de los datos locales y detecta valores atípicos mediante el cálculo basado en LOF. Define una nueva medida llamada Factor de valor atípico basado en conglomerados (CBOF), que captura la esencia de los valores atípicos basados en conglomerados. Cuanto mayor sea el CBOF de un grupo, más anormal será ese grupo.


### Esquema de aprendizaje

Existen algoritmos de detección de valores atípicos tanto de variedades predictivas (supervisadas) como directas (no supervisadas). Al observar varias formas del problema de detección de valores atípicos en diversos entornos de aplicación, es posible agrupar varias metodologías en tres categorías amplias según el tipo de esquema de aprendizaje involucrado.

1. **Determinar valores atípicos sin conocimiento previo de los datos**: Este es esencialmente un enfoque de aprendizaje análogo al aprendizaje no supervisado (clustering). Este enfoque considera los datos como una distribución estática y marca los puntos más remotos como posibles valores atípicos.

2. **Modelación tanto de la normalidad como la anormalidad**: este enfoque es análogo al aprendizaje supervisado y requiere datos preetiquetados correspondientes a clases normales y anormales.

3. **Modelar solo la normalidad o, en algunos casos, modelar la anormalidad**: Este enfoque se conoce generalmente como detección de novedad. Es análogo a una detección semisupervisada, ya que se enseña la clase normal y el algoritmo aprende a reconocer la anomalía por sí mismo.

El método de detección también influye en el tipo de técnica de aprendizaje a considerar. Mientras que las medidas supervisadas determinan la similitud en función de la información de clase, las medidas basadas en datos determinan la similitud
basado en la distribución de datos. En principio, ambas ideas se pueden combinar para desarrollar métodos híbridos.


### Técnicas de aprendizaje supervisado

Las redes neuronales y los árboles de decisión son dos formas comunes de análisis predictivo de valores atípicos. Como ya se sabe, las redes neuronales supervisadas procesan datos numéricos y los métodos basados en árboles de decisión tratan con datos categóricos.

Los métodos de aprendizaje supervisado generalmente construyen un modelo de predicción para objetos raros basados en datos etiquetados y lo usan para clasificar cada objeto como un objeto atípico o normal.

Estos enfoques enfrentan el problema de la distribución de clases desequilibrada
en los datos, lo que a su vez puede afectar el rendimiento del modelo de clasificación subyacente. 
Las principales desventajas de las técnicas supervisadas incluyen la necesidad de tener datos etiquetados y la incapacidad de detectar nuevas clases de objetos raros.
